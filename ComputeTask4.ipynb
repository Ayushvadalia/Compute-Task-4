```python
#fit the model
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
import numpy as np

# Assuming X_train is your 1D array
X_train = np.array(X_train)

# Add a new axis to convert it to a 2D array
X_train_2d = X_train[:, np.newaxis]
#[:, np.newaxis]: This is a NumPy slicing operation. The colon (:) indicates that we want to include
# all elements along the first axis (rows).
 #The np.newaxis is used to add a new axis along the second axis (columns).
```

```python
import numpy as np

# Assuming X_train is your 1D array
X_test = np.array(X_test)

# Add a new axis to convert it to a 2D array
X_test_2d = X_test[:, np.newaxis]
```

```python
from sklearn.linear_model import LinearRegression

# Create a linear regression model
model = LinearRegression()

# Assuming you have training data X_train and corresponding labels y_train
# Train the model
model.fit(X_train_2d, y_train)

# Make predictions on new data (X_test)
y_pred = model.predict(X_test_2d)
```

## EVALUATING THE MODEL

```python
from sklearn.metrics import mean_absolute_error
print("MAE :",mean_absolute_error(y_test,y_pred))
#mean absolute error
```

```python
from sklearn.metrics import mean_squared_error
print("MSE",mean_squared_error(y_test,y_pred))
#mean square error
```

```python
print("RMSE",np.sqrt(mean_squared_error(y_test,y_pred)))
#root mean square error
```

```python
print("RMSLE",np.log(np.sqrt(mean_squared_error(y_test,y_pred))))
#root mean square log error
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test,y_pred)
print(r2)
#r2 square
```
